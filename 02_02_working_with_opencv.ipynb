## 0. Opening an Image File


```python
import cv2
from matplotlib import pyplot as plt

image_file = 'data/page_01.png'
img = cv2.imread(image_file)
# cv2.imshow("Original Image: ", img)
# cv2.waitKey(0)
```

```python
# Display image inline

import matplotlib.pyplot as plt

def display(im_path):  # image path: 'data/page_01.png'
    dpi = 80
    im_data = plt.imread(im_path)

    # Handle grayscale and color images
    if im_data.ndim == 2:
        height, width = im_data.shape
        depth = 1
    elif im_data.ndim == 3:
        height, width, depth = im_data.shape
    else:
        raise ValueError("Unsupported image format: expected 2D or 3D image")

    # Size of figure in inches
    figsize = width / float(dpi), height / float(dpi)

    # Create figure and axes
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.axis('off')  # remove ticks and spines

    # Display image with appropriate colormap
    if depth == 1:
        ax.imshow(im_data, cmap='gray')
    else:
        ax.imshow(im_data)

    plt.show()

    
display(image_file)
```

## 1. Invert Image - Switch colors

```python
inverted_image = cv2.bitwise_not(img) # use loaded memeory, not path
cv2.imwrite("temp/inverted.png", inverted_image)

display("temp/inverted.png")
```

## 03. Binarization

```python
# Convert an image to black and white

# function to get grayscale
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# for og image
gray_image = grayscale(img)
cv2.imwrite("temp/gray.png", gray_image)
display("temp/gray.png")

# for inverted image
gray_image = grayscale(inverted_image)
cv2.imwrite("temp/gray1.png", gray_image)
display("temp/gray1.png")
```

```python
import cv2

# Read the image
img = cv2.imread("data/handwritten_sample.jpeg")

# Upscale by 2x
upscaled = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Save the upscaled image
cv2.imwrite("temp/upscaled_handwritten_sample.jpeg", upscaled)

display("temp/upscaled_handwritten_sample.jpeg")
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread("temp/upscaled_handwritten_sample.jpeg")

# Sharpening kernel
kernel = np.array([[0, -1, 0],
                [-1, 5, -1],
                [0, -1, 0]]
                )   

# Apply the sharpening kernel
sharpened = cv2.filter2D(img, -1, kernel)

# Save and optionally display
cv2.imwrite("temp/sharpened_handwritten_sample.jpeg", sharpened)

display("temp/sharpened_handwritten_sample.jpeg")
```
